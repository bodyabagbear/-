Завдання 1. Метод прямокутників


```python

from scipy import integrate
import numpy as np

def left_rec(f, a, b, n):
    h = (b - a) / n
    sum_val = 0
    for i in range(0, n):
        sum_val += f(a + i * h)
    return sum_val * h

def right_rec(f, a, b, n):
    h = (b - a) / n
    sum_val = 0
    for i in range(1, n + 1):
        sum_val += f(a + i * h)
    return sum_val * h

def aver_rec(f, a, b, n):
    h = (b - a) / n
    sum_val = 0
    for i in range(0, n):
        sum_val += f(a + (i + 0.5) * h)
    return sum_val * h

def simpson_rule(f, a, b, n):
    h = (b - a) / n
    integr = f(a) + f(b)
    for i in range(1, n):
        k = a + i * h
        if i % 2 == 0:
            integr += 2 * f(k)
        else:
            integr += 4 * f(k)
    integr *= h / 3
    return integr

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    sum_trap = 0.5 * (f(a) + f(b))
    x = a
    for i in range(1, n):
        x += h
        sum_trap += f(x)
    return h * sum_trap

print("Інтеграл 1 (Метод прямокутників)")
n1 = 10
a1, b1 = 0.15, 0.5

def f1(x):
    return 1 / np.sqrt(2 * x + 1.6)

print("Метод лівих прямокутників:   ", round(left_rec(f1, a1, b1, n1), 4))
print("Метод правих прямокутників:  ", round(right_rec(f1, a1, b1, n1), 4))
print("Метод середніх прямокутників:", round(aver_rec(f1, a1, b1, n1), 4))
v1, err1 = integrate.quad(f1, a1, b1)
print("Перевірка через SciPy:       ", round(v1, 4))
print()


print("Інтеграл 2 (Метод Сімпсона)")
n2 = 8
a2, b2 = 0.5, 1.2

def f2(x):
    return np.tan(x**2) / (x + 1)

print("Метод Сімпсона:              ", round(simpson_rule(f2, a2, b2, n2), 4))
v2, err2 = integrate.quad(f2, a2, b2)
print("Перевірка через SciPy:       ", round(v2, 4))
print()


print("Інтеграл 3 (Метод трапецій)")
n3 = 20
a3, b3 = 2.0, 3.5

def f3(x):
    return 1 / np.sqrt(x**2 - 1)

print("Метод трапецій:              ", round(trapezoidal_rule(f3, a3, b3, n3), 4))
v3, err3 = integrate.quad(f3, a3, b3)
print("Перевірка через SciPy:       ", round(v3, 4))
```

Завдання 2. Метод Сімпсона


```python

from scipy import integrate
import numpy as np

def simpson_rule(f, a, b, n):
    if n % 2 != 0:
        n += 1

    h = (b - a) / n
    integr = f(a) + f(b)

    for i in range(1, n):
        k = a + i * h
        if i % 2 == 0:
            integr += 2 * f(k)
        else:
            integr += 4 * f(k)

    integr *= h / 3
    return integr

n = 8

print("Інтеграл 1")
a1, b1 = 0.15, 0.5

def f1(x):
    return 1 / np.sqrt(2 * x + 1.6)

print("Simpson method:  ", round(simpson_rule(f1, a1, b1, n), 4))
v1, err1 = integrate.quad(f1, a1, b1)
print("SciPy quad check:", round(v1, 4))
print()

print("Інтеграл 2")
a2, b2 = 0.5, 1.2

def f2(x):
    return np.tan(x**2) / (x + 1)

print("Simpson method:  ", round(simpson_rule(f2, a2, b2, n), 4))
v2, err2 = integrate.quad(f2, a2, b2)
print("SciPy quad check:", round(v2, 4))
print()

print("Інтеграл 3")
a3, b3 = 2.0, 3.5

def f3(x):
    return 1 / np.sqrt(x**2 - 1)

print("Simpson method:  ", round(simpson_rule(f3, a3, b3, n), 4))
v3, err3 = integrate.quad(f3, a3, b3)
print("SciPy quad check:", round(v3, 4))
```

Завдання 3. Визначити інтеграм методом трапецій

```python

from scipy import integrate
import numpy as np

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    sum_trap = 0.5 * (f(a) + f(b))
    x = a

    for i in range(1, n):
        x += h
        sum_trap += f(x)

    return h * sum_trap

n = 20

print("Інтеграл 1")
a1, b1 = 0.15, 0.5

def f1(x):
    return 1 / np.sqrt(2 * x + 1.6)

print("Trapezoidal method:", round(trapezoidal_rule(f1, a1, b1, n), 4))
v1, err1 = integrate.quad(f1, a1, b1)
print("SciPy quad check:  ", round(v1, 4))
print()

print("Інтеграл 2")
a2, b2 = 0.5, 1.2

def f2(x):
    return np.tan(x**2) / (x + 1)

print("Trapezoidal method:", round(trapezoidal_rule(f2, a2, b2, n), 4))
v2, err2 = integrate.quad(f2, a2, b2)
print("SciPy quad check:  ", round(v2, 4))
print()

print("Інтеграл 3")
a3, b3 = 2.0, 3.5

def f3(x):
    return 1 / np.sqrt(x**2 - 1)

print("Trapezoidal method:", round(trapezoidal_rule(f3, a3, b3, n), 4))
v3, err3 = integrate.quad(f3, a3, b3)
print("SciPy quad check:  ", round(v3, 4))
```
